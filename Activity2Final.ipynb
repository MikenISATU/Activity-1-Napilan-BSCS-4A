```python
#3 Import SimpleImputer library for missing data

from sklearn.impute import SimpleImputer
from pandas import read_csv #but used df as it was already inputted to the df variable name.
from numpy import nan
from numpy import isnan

df = pd.read_csv('Data-customer.csv')

imputer1 = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer2 = SimpleImputer(missing_values=np.nan, strategy='most_frequent')


df[['Age', 'Salary']] = imputer1.fit_transform(df[['Age','Salary']])
df[['Status', 'Employment']] = imputer2.fit_transform(df[['Status', 'Employment']])

dt = df

# Missing Values are filled
print(df)
```

```python
# Encoding Categorical Values
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Purchased"] = label_encoder.fit_transform(df['Purchased']) #Dependent

# Do encode Independent Variable
df_encoded = pd.get_dummies(df, columns=['Location', 'Status', 'Employment','Age','Salary'], drop_first=True)
df_encoded
```

```python
# Do split the dataset to the training and test set
from sklearn.model_selection import train_test_split

x = df_encoded.drop('Purchased', axis=1)
y = df_encoded['Purchased']
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=42)
```
